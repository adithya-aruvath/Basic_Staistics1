#### Bar Chart Analysis for Categorical Columns:

```python
for col in cat_col:
    freq=df[col].value_counts().sort_index()
    plt.figure(figsize=(6,4))
    plt.bar(freq.index,freq.values,color='brown')
    plt.title(f'Frequency of Categories in {col}')
    plt.xlabel(col)
    plt.ylabel('frequency')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

### Standardization of Numerical Variables

```python
## Standardization is a technique used to scale numerical variables so that they have a mean of 0 and a standard deviation of 1.
## The standardized value (z-score) is calculated using the formula,
## 𝑧 =(𝑋−𝜇)/𝜎
##X is the original value, μ is the mean of the variable, σ is the standard deviation of the variable.
```

```python
standardized_vol=(df['Volume']-df['Volume'].mean())/df['Volume'].std()
```

```python
standardized_vol
```

```python
standardized_avgprice=(df['Avg Price']-df['Avg Price'].mean())/df['Avg Price'].std()
standardized_avgprice
```

```python
standardized_totalsalesvalue=(df['Total Sales Value']-df['Total Sales Value'].mean())/df['Total Sales Value'].std()
standardized_totalsalesvalue
```

```python
standardized_discountrate=(df['Discount Rate (%)']-df['Discount Rate (%)'].mean())/df['Discount Rate (%)'].std()
standardized_discountrate
```

```python
standardized_discountamount=(df['Discount Amount']-df['Discount Amount'].mean())/df['Discount Amount'].std()
standardized_discountamount
```

```python
standardized_netsalesvalue=(df['Net Sales Value']-df['Net Sales Value'].mean())/df['Net Sales Value'].std()
standardized_netsalesvalue
```

```python
sns.histplot(standardized_vol, bins=50, kde=True, color='blue', label='Standardized Volume')
plt.title('Standardized Volume')
plt.xlabel('Z-score Values')
plt.ylabel('Frequency')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
sns.histplot(standardized_avgprice, bins=50, kde=True, color='blue', label='Standardized Avg Price')
plt.title('Standardized Avg Price')
plt.xlabel('Z-score Values')
plt.ylabel('Frequency')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
sns.histplot(standardized_totalsalesvalue, bins=50, kde=True, color='blue', label='Standardized Total Sales Value')
plt.title('Standardized Total Sales Value')
plt.xlabel('Z-score Values')
plt.ylabel('Frequency')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
sns.histplot(standardized_discountrate, bins=50, kde=True, color='blue', label='Standardized Discount Rate')
plt.title('Standardized Discount Rate')
plt.xlabel('Z-score Values')
plt.ylabel('Frequency')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
sns.histplot(standardized_discountamount, bins=50, kde=True, color='blue', label='Standardized Discount Amount')
plt.title('Standardized Discount Amount')
plt.xlabel('Z-score Values')
plt.ylabel('Frequency')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
sns.histplot(standardized_netsalesvalue, bins=50, kde=True, color='blue', label='Standardized Net Sales Value')
plt.title('Standardized Net Sales Value')
plt.xlabel('Z-score Values')
plt.ylabel('Frequency')
plt.legend()
plt.tight_layout()
plt.show()
```
